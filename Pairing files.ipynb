```python
def stratified_split(df, stratify_col, test_size=0.3, val_size=0.1):
    # Split df into train + validation and test sets
    train_val_df, test_df = train_test_split(
        df, 
        test_size=test_size, 
        stratify=df[stratify_col]
    )

    # Split train + validation into train and validation sets
    train_df, val_df = train_test_split(
        train_val_df, 
        test_size=val_size / (1 - test_size), 
        stratify=train_val_df[stratify_col]
    )

    return train_df, test_df, val_df

train_df, test_df, val_df = stratified_split(merged_ucla, stratify_col='speakerID')
```

```python
from itertools import combinations
import random

def pairing_algorithm(df, n, seed, filter_criteria, nflag=False):
    speakers = np.unique(df['speakerID'].values)
    random.seed = seed
    d_same = []
    d_diff = []
    for speaker in speakers:
        utterances_same = df[df['speakerID'] == speaker]['path'].tolist()
        utterances_diff = df[df['speakerID'] != speaker]['path'].tolist()
        utterances_diff_all = df[df['speakerID'] != speaker]['path'].tolist()
        if filter_criteria != []:
            crit1 = df[df['speakerID'] == speaker][filter_criteria[0]].tolist()[0]
            crit2 = df[df['speakerID'] == speaker][filter_criteria[1]].tolist()[0]

            # Filter the DataFrame based on the filter criteria
            filtered_df = df[(df[filter_criteria[0]] == crit1) & (df[filter_criteria[1]] == crit2) & (df['speakerID'] != speaker)]  # Add more conditions if needed
            utterances_diff = filtered_df['path'].tolist()
            utterances_diff_all = list(set(utterances_diff_all) - set(utterances_diff))

        same_pairs = []
        diff_pairs = []
        # Make all pairs with same speaker
        same_pairs.extend(list(combinations(utterances_same, 2)))
        # Sample n random pairs
        if nflag == False:
            n = len(same_pairs)
        if n > len(utterances_diff):
            diff = n - len(utterances_diff)
            rand = random.sample(utterances_diff_all, diff)
            utterances_diff.extend(rand)
        else:
            utterances_diff = random.sample(utterances_diff, n)
        same_pairs = random.sample(same_pairs, n)
        utterances_same = random.choices(utterances_same, k=n)
#         print(utterances_same)
        x = 0
        for same, diff in zip(utterances_same, utterances_diff):
            diff_pairs.append((same, diff))
            
        d_same.extend(same_pairs)
        d_diff.extend(diff_pairs)
    combined_list = [(pair[0], pair[1], True) for pair in d_same] + [(pair[0], pair[1], False) for pair in d_diff]
    df = pd.DataFrame(combined_list, columns=['file1', 'file2', 'label'])

    return df      
filter_criteria = ['sex', 'L2=EnglishL1'] 
df = pairing_algorithm(merged_ucla, 3, 52, filter_criteria)
df
            
        
```

```python
def sample_group(group, n):
    return group.sample(min(n, len(group)), replace=True)

def sample_df(df, label_column, n):
    groups = df.groupby(label_column)
    sampled_groups = groups.apply(lambda x: sample_group(x, n // len(groups)))
    return sampled_groups.sample(n)
sampled_df = sample_df(df, 'label', 5000)
```

```python
import os
import pandas as pd

# Define the directory where your folders are located
folder_path = "Wyred/Cut_Wyred"

# Initialize an empty list to store file attributes
file_attributes = []

# Iterate through each folder
for folder_name in os.listdir(folder_path):
    folder_dir = os.path.join(folder_path, folder_name)
    
    # Iterate through each file in the folder
    for filename in os.listdir(folder_dir):
        file_info = filename.split('-')
        
        # Extracting information from the filename
        speaker_id = file_info[0]
        
        # Handling Task 2 specific round information
        if len(file_info) == 6:  # Task 2 filename
            round_info = file_info[1]
            task_number = file_info[2]
            place = file_info[3]
            quality = file_info[4]
        else:
            round_info = None
            task_number = file_info[1]
            place = file_info[2]
            quality = file_info[3]
        
        # Handling Task 3 specific information
        if task_number == '3':
            friend_or_nofriend = file_info[4]
            matched_id = file_info[5]
        else:
            friend_or_nofriend = None
            matched_id = None
        
        date = file_info[-1].split('.')[0] # Removing file extension
        
        # Creating filepath
        filepath = os.path.join(folder_dir, filename)
        
        # Append the file attributes to the list
        file_attributes.append([speaker_id, round_info, task_number, place, quality, friend_or_nofriend, matched_id, date, filepath])

# Create a pandas DataFrame
columns = ['speakerID', 'Round', 'TaskNumber', 'Place', 'Quality', 'FriendOrNofriend', 'MatchedID', 'Date', 'path']
df = pd.DataFrame(file_attributes, columns=columns)


# Display the DataFrame
df
```
